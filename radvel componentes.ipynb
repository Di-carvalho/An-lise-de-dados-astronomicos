# Espectroscopia

# Hb

```python
# Lendo o arquivo.

radvel = pd.read_csv('componentes_hb.txt', sep = ' ', index_col = False)
#radvel
```

```python
# Calcula as velocidades radiais.

from astropy.constants import c

radvel['radvel_broad']  = ((radvel['broad'] - 4861.35)/4861.35)*(c.to('km/s'))
radvel['radvel_narrow'] = ((radvel['narrow'] - 4861.35)/4861.35)*(c.to('km/s'))

#radvel.sort_values(by = 'fase')
```

```python
# Coloca o bjd no arquivo.

radvel['time'] = pd.read_csv('hjd_espectros.txt')
```

```python
# Calculo da fase

radvel['fase'] = ((radvel['time'] -2456980.6726) /0.07918 ) % 1

# Duplica a fase.

radvel1 = radvel.copy()
radvel1['fase'] = radvel1['fase'] + 1
radvel = pd.concat([radvel, radvel1], ignore_index=True)

#radvel.sort_values('fase')
```

```python
#Plota o gráfico das velocidades radiais.

f, ax = plt.subplots(figsize=(12,7))

plt.scatter(radvel['fase'], radvel['radvel_broad'], color = 'blue',s = 5)
plt.scatter(radvel['fase'], radvel['radvel_narrow'], color = 'red', s = 5)
```

```python
# Plota os gráficos do fluxo.

f, ax = plt.subplots(figsize=(12,7))

radvel = radvel.sort_values(by = 'fase')

plt.plot(radvel['fase'], radvel['fluxb'], color = 'blue')
plt.plot(radvel['fase'], radvel['fluxn'], color = 'red')
```

```python
# Plota os Gráficos da largura a meia altura.

f, ax = plt.subplots(figsize=(12,7))


plt.plot(radvel['fase'], radvel['fwhmb'], color = 'blue')
plt.plot(radvel['fase'], radvel['fwhmn'], color = 'red')
```

```python
# Plota os gráficos da largura equivalente.

f, ax = plt.subplots(figsize=(12,7))


plt.plot(radvel['fase'], radvel['equb'], color = 'blue')
plt.plot(radvel['fase'], radvel['equn'], color = 'red')
```

```python
radvel_hb = radvel.drop(columns = ['broad', 'fluxb', 'equb', 'fwhmb', 'narrow', 'fluxn', 'equn', 'fwhmn',
                                   'time'])
flux_hb = radvel.drop(columns = ['broad', 'equb', 'fwhmb', 'narrow', 'equn', 'fwhmn', 'radvel_broad', 'radvel_narrow',
                                   'time'])
equ_hb = radvel.drop(columns = ['broad', 'fluxb', 'fwhmb', 'narrow', 'fluxn', 'fwhmn', 'radvel_broad', 'radvel_narrow',
                                   'time'])
fwhmb_hb = radvel.drop(columns = ['broad', 'fluxb', 'equb', 'narrow', 'fluxn', 'equn', 'radvel_broad', 'radvel_narrow',
                                   'time'])
```

```python
radvel_hb.to_csv('radvel_hb.txt')
flux_hb.to_csv('flux_hb.txt')
equ_hb.to_csv('equ_hb.txt')
fwhmb_hb.to_csv('fwhmb_hb.txt')
```

## Busca de períodos.

```python
circpolI = pd.read_csv('circpolI_bjd_semerro.txt', sep = ' ')
```

```python
from astropy.timeseries.periodograms.lombscargle import LombScargle

freq, PLS = LombScargle(circpolI['bjd'], circpolI['circ_pol']).autopower(minimum_frequency=5 ,maximum_frequency=20, method='chi2', samples_per_peak=20)
best_freq = freq[np.argmax(PLS)]
print('melhor freq = ',best_freq)
print('PLS max = ',PLS.max())
print('FAP =',LombScargle(circpolI['bjd'], circpolI['bjd']).false_alarm_probability(PLS.max()))
```

```python
fap50 = LombScargle(circpolI['bjd'], circpolI['circ_pol']).false_alarm_level(0.5)
fap10 = LombScargle(circpolI['bjd'], circpolI['circ_pol']).false_alarm_level(0.1)
fap1 = LombScargle(circpolI['bjd'], circpolI['circ_pol']).false_alarm_level(0.01)
print('Fap 50% = ', fap50)
print('Fap 10% = ', fap10)
print('Fap 1% = ', fap1)
```

```python
#plt.vlines(best_freq,0,0.012,color='k',linestyles='dashed')
plt.plot(freq,PLS)
plt.xlabel("Frequência (Ciclos por Dia. )")
plt.ylabel("Intensidade Lomb-Scargle")
plt.savefig('teste.png', format='png')
plt.hlines(fap50,5,20,color='b',linestyles='dashed')
plt.hlines(fap10,5,20,color='r',linestyles='dashed')
plt.hlines(fap1,5,20,color='g',linestyles='dashed')
plt.vlines(best_freq,0,0.6,color='y',linestyles='dashed')


plt.show()
lb=pd.DataFrame()
lb['freq']=freq
lb['pls']= PLS
lb.to_csv('lb_filtroI.txt',index=False,header=False,sep=' ' )
```
